```python
import pandas as pd

# Load the dataset
file_path = "merged_ecg_data_cleaned.csv"
df = pd.read_csv(file_path)

# Compute summary statistics
table = {
    "Feature": ["Total Records", "Unique Subjects", "Unique Studies", "Unique Carts", "Avg RR Interval", "Avg QRS Duration", "Avg P Onset", "Avg P End", "Avg QRS Onset", "Avg QRS End", "Avg T End", "Avg P Axis", "Avg QRS Axis", "Avg T Axis", "Total WCT Cases"],
    "Value": [
        len(df),  # Total number of records
        df["subject_id"].nunique(),  # Unique subjects
        df["study_id"].nunique(),  # Unique studies
        df["cart_id"].nunique(),  # Unique carts
        df["rr_interval"].mean(),  # Average RR Interval
        df["qrs_duration"].mean(),  # Average QRS Duration
        df["p_onset"].mean(),  # Average P Onset
        df["p_end"].mean(),  # Average P End
        df["qrs_onset"].mean(),  # Average QRS Onset
        df["qrs_end"].mean(),  # Average QRS End
        df["t_end"].mean(),  # Average T End
        df["p_axis"].mean(),  # Average P Axis
        df["qrs_axis"].mean(),  # Average QRS Axis
        df["t_axis"].mean(),  # Average T Axis
        df["wct_label_encoded"].sum()  # Total WCT Cases
    ]
}

# Convert to DataFrame and display
stats_df = pd.DataFrame(table)
print(stats_df)

# Save to CSV for future use
stats_df.to_csv("ecg_statistics_summary.csv", index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "merged_ecg_data_cleaned.csv"
df = pd.read_csv(file_path)

# Compute summary statistics
table = {
    "Feature": ["Total Records", "Unique Subjects", "Unique Studies", "Unique Carts", "Avg RR Interval", "Avg QRS Duration", "Avg P Onset", "Avg P End", "Avg QRS Onset", "Avg QRS End", "Avg T End", "Avg P Axis", "Avg QRS Axis", "Avg T Axis", "Total WCT Cases"],
    "Value": [
        len(df),  # Total number of records
        df["subject_id"].nunique(),  # Unique subjects
        df["study_id"].nunique(),  # Unique studies
        df["cart_id"].nunique(),  # Unique carts
        df["rr_interval"].mean(),  # Average RR Interval
        df["qrs_duration"].mean(),  # Average QRS Duration
        df["p_onset"].mean(),  # Average P Onset
        df["p_end"].mean(),  # Average P End
        df["qrs_onset"].mean(),  # Average QRS Onset
        df["qrs_end"].mean(),  # Average QRS End
        df["t_end"].mean(),  # Average T End
        df["p_axis"].mean(),  # Average P Axis
        df["qrs_axis"].mean(),  # Average QRS Axis
        df["t_axis"].mean(),  # Average T Axis
        df["wct_label_encoded"].sum()  # Total WCT Cases
    ]
}

# Convert to DataFrame and display
stats_df = pd.DataFrame(table)
print(stats_df)

# Set the font to Times New Roman
plt.rcParams['font.family'] = 'Times New Roman'

# Generate line chart for statistics with high quality
plt.figure(figsize=(12, 8), dpi=300)  # Increased resolution for high-quality image
plt.plot(stats_df["Feature"], stats_df["Value"], marker='o', color='teal', linestyle='-', linewidth=2, markersize=6)  # Line plot with markers
plt.xlabel("Feature", fontsize=14)  # Increase font size for axis label
plt.ylabel("Value", fontsize=14)  # Increase font size for axis label
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for readability
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Keep the grid for better readability
plt.tight_layout()  # Ensures the plot fits within the figure without clipping

# Save the figure locally
output_path = "ecg_statistics_line_chart.png"  # Specify the output file path
plt.savefig(output_path, format='png', dpi=300)  # Save with high quality (300 dpi)

# Optionally, display the figure
plt.show()

print(f"Chart saved as {output_path}")
```
