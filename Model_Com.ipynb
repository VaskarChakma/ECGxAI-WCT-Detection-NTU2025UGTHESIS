```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os
from matplotlib import rcParams

# Set Times New Roman globally
plt.rcParams["font.family"] = "Times New Roman"
rcParams.update({'font.size': 12})

# Create directory if not exists
os.makedirs("NewExp", exist_ok=True)

# Dataset details
total_samples = 300000
positive = int(total_samples * 0.515)
negative = total_samples - positive

# Provided performance data
performance_data = {
    'XGBoost': {'accuracy': 0.6656, 'f1': 0.6454},
    'LightGBM': {'accuracy': 0.6532, 'f1': 0.6396},
    'Random Forest': {'accuracy': 0.6946, 'f1': 0.6589},
    'Gradient Boosting': {'accuracy': 0.6368, 'f1': 0.6280}
}

# Matching the color palette used in the original image
color_palettes = {
    'XGBoost': "Blues",
    'LightGBM': "Reds",
    'Random Forest': "Greens",
    'Gradient Boosting': "Reds"
}

# Generate confusion matrix estimates
def generate_conf_matrix(accuracy, f1_score, pos, neg):
    tp = int(f1_score * pos)
    fn = pos - tp
    tn_fp = total_samples - tp - fn
    tn = int(accuracy * total_samples - tp)
    fp = tn_fp - tn
    return [[tn, fp], [fn, tp]]

# Generate confusion matrices
conf_matrices = {
    model: generate_conf_matrix(metrics['accuracy'], metrics['f1'], positive, negative)
    for model, metrics in performance_data.items()
}

# Plot
fig, axes = plt.subplots(2, 2, figsize=(14, 12))
axes = axes.flatten()

for ax, (model, matrix) in zip(axes, conf_matrices.items()):
    matrix = np.array(matrix)
    tn, fp = matrix[0]
    fn, tp = matrix[1]
    sensitivity = tp / (tp + fn) if (tp + fn) > 0 else 0
    specificity = tn / (tn + fp) if (tn + fp) > 0 else 0

    cmap = sns.color_palette(color_palettes[model], as_cmap=True)

    sns.heatmap(matrix, annot=True, fmt="d", cmap=cmap, ax=ax, cbar=False,
                xticklabels=["Negative", "Positive"],
                yticklabels=["Negative", "Positive"],
                linewidths=0.5, linecolor='black', annot_kws={"fontsize": 12})

    ax.set_title(f"{model}\nSens: {sensitivity:.3f}, Spec: {specificity:.3f}", fontsize=14)
    ax.set_xlabel("Predicted Label", fontsize=12)
    ax.set_ylabel("True Label", fontsize=12)

plt.tight_layout()
plt.savefig("NewExp/Confusion metrics.png", dpi=300)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import os
from sklearn.metrics import mean_squared_error

# Create NewExp directory if it doesn't exist
os.makedirs('NewExp', exist_ok=True)

# Set style
plt.style.use('seaborn-v0_8')
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['figure.dpi'] = 300

# Generate synthetic data based on your simulation results
np.random.seed(42)
x = np.linspace(0, 1, 20)
y_true = 0.5 * np.sin(2 * np.pi * x) - 0.2  # Adjusted to fit your y-axis range
y_noise = y_true + np.random.normal(0, 0.05, len(x))  # Less noise for better visualization

# Create polynomial fits using your model metrics as inspiration
underfit_coeff = np.polyfit(x, y_noise, 1)  # Linear (like XGBoost's lower performance)
goodfit_coeff = np.polyfit(x, y_noise, 3)   # Cubic (like RandomForest's good performance)
overfit_coeff = np.polyfit(x, y_noise, 12)  # High degree (like over-optimized models)

x_plot = np.linspace(0, 1, 100)
underfit = np.polyval(underfit_coeff, x_plot)
goodfit = np.polyval(goodfit_coeff, x_plot)
overfit = np.polyval(overfit_coeff, x_plot)

# Calculate RMSE for each fit (compatible with all sklearn versions)
rmse_underfit = np.sqrt(mean_squared_error(y_true, np.polyval(underfit_coeff, x)))
rmse_goodfit = np.sqrt(mean_squared_error(y_true, np.polyval(goodfit_coeff, x)))
rmse_overfit = np.sqrt(mean_squared_error(y_true, np.polyval(overfit_coeff, x)))

# Create figure with adjusted size
plt.figure(figsize=(10, 6))

# Plot elements with adjusted styles
plt.scatter(x, y_noise, color='black', s=60, label='Data points', zorder=3)
plt.plot(x_plot, underfit, 'r--', linewidth=3, 
         label=f'Underfitting (RMSE: {rmse_underfit:.3f})')
plt.plot(x_plot, goodfit, 'g-', linewidth=3, 
         label=f'Good fit (RMSE: {rmse_goodfit:.3f})')
plt.plot(x_plot, overfit, 'b:', linewidth=3, 
         label=f'Overfitting (RMSE: {rmse_overfit:.3f})')

# Customize plot to perfectly fit curves
plt.title('Model Fit Comparison with Performance Metrics', fontsize=14, pad=15)
plt.xlabel('Input (Simulation Number)', fontsize=12, labelpad=10)
plt.ylabel('Output (RMSE)', fontsize=12, labelpad=10)
plt.ylim(-1.0, 0.5)  # Matches your reference y-axis
plt.xlim(-0.05, 1.05)  # Slightly extended to ensure curves fit
plt.grid(True, alpha=0.3)

# Add legend with metrics from your data
plt.legend(loc='upper right', fontsize=10, 
           title='Model Fits with RMSE', 
           title_fontsize=11)

# Add annotation box with your metrics summary
metrics_text = ("RMSE Metrics:\n"
                "XGBoost: 1.0001\n"
                "LightGBM: 0.9992\n"
                "RandomForest: 0.9959\n"
                "GradientBoost: 1.0485")
plt.text(0.02, -0.9, metrics_text, fontsize=10,
         bbox=dict(facecolor='white', alpha=0.8))

# Adjust layout to ensure everything fits
plt.tight_layout()

# Save and show
plt.savefig('NewExp/model_fit_comparison.png', bbox_inches='tight', dpi=300)
plt.show()

print("Plot saved to NewExp/model_fit_comparison.png")
```
